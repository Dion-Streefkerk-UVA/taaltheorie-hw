### **Section 1: Context-Free Grammars**

In this set of exercises you are asked to implement several grammars using the Python NLTK library. To do this you will use some classes which have been imported at the beginning of this notebook.

First, we will define a function `check_sentence` which will take a sentence and a grammar as input and will return true if the sentence is in the language of the grammar and false otherwise. This function will be used to test the grammars you will implement in the following exercises.

```python
def check_sentence(parser, sentence):
    """
    Check if a sentence is valid according to the provided grammar.
    """

    # If the sentence is a string, split it into a list of words
    if isinstance(sentence, str):
        sentence = sentence.split()

    # Check if the sentence is valid
    tree_found = False

    # Parse the sentence
    results = parser.parse(sentence)

    # Loop through the results
    for tree in results:
        tree_found = True

    return tree_found
```
